```python
# we have to correct some company names
cars.CompanyName = cars.CompanyName.str.lower() # convert all company names to lower case
def replace_name(a,b):
    cars.CompanyName.replace(a, b, inplace=True) # replace the old name with the new name

replace_name('maxda', 'mazda')
replace_name('porcshce', 'porsche')
replace_name('toyouta', 'toyota')
replace_name('vokswagen', 'volkswagen')
replace_name('vw', 'volkswagen')
replace_name('chevroelt', 'chevrolet')

cars.CompanyName.unique() # check the unique values again
```

FEATURE ENGINEERING

```python
# Derive new column - fuel economy
cars['fuel_economy']= (0.55 * cars['citympg']) + (0.45 * cars['highwaympg']) # giving more weigtage to city milage is 55 % and for highway milage is 45 %
```

```python
cars['price'] = cars['price'].astype(int) # convert price column to int type
temp = cars.copy() # create a copy of the dataframe
```

```python
temp.groupby(['CompanyName'])['price'].mean()
```

```python
table = temp.groupby(['CompanyName'])['price'].mean() # group by CompanyName and calculate the mean price
temp = temp.merge(table.reset_index(),how= 'left',on= 'CompanyName') # merge the mean price with the original dataframe using left join
```

```python
# have to derive new column - cartype (Budget , Medium, Highrange)
bins =[0,10000,20000,40000] # it is 0- 9999, 10000-19999, 20000-39999
cars_bin = ['Budget', 'Medium', 'Highend'] # define the bins for the price rang
cars['carsrange'] = pd.cut(temp['price_y'],bins,right=False, labels=cars_bin) # create bins for the price range
cars.head() # check the new column
```

```python
# converting categorical columns to numerical columns
def dummies(x,df):
    temp = pd.get_dummies(df[x],drop_first=True, dtype=float) # create dummy variables for the categorical column, drop_first=True will drop the first category to avoid dummy variable trap
    df = pd.concat([df,temp],axis=1) # concatenate the dummy variables with the original dataframe
    df.drop([x], axis=1, inplace=True) # drop the original categorical column
    return df
```

```python
# Applying the dummies function to the cars_lr
cars_lr = cars
cars_lr = dummies('fueltype', cars_lr)
cars_lr = dummies('aspiration', cars_lr)
cars_lr = dummies('carbody', cars_lr)
cars_lr = dummies('drivewheel', cars_lr)
cars_lr = dummies('enginetype', cars_lr)
cars_lr = dummies('cylindernumber', cars_lr)
cars_lr = dummies('carsrange', cars_lr)
cars_lr = dummies('CompanyName', cars_lr) # convert the CompanyName column to dummy variables
cars_lr = dummies('doornumber', cars_lr)
cars_lr = dummies('enginelocation', cars_lr)
cars_lr = dummies('fuelsystem', cars_lr) 
```

```python
from sklearn.preprocessing import MinMaxScaler # to scale the numerical columns to a range between 0 and 1

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'carheight', 'stroke', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fuel_economy','carlength','carwidth','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars]) 
```

BUILDING MODEL

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def build_model(X,y):
    X = sm.add_constant(X) # Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary())    # model summary
    return lm
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

### Feature Elimination using RFE (Recursive Feature Elimination)
Recursive feature elimination (RFE) is a feature selection method that fits a model and removes the weakest feature (or features) until the specified number of features is reached. Features are ranked by the model’s coef_ or feature_importances_ attributes, and by recursively eliminating a small number of features per loop, RFE attempts to eliminate dependencies and collinearity that may exist in the model.

```python
lm = LinearRegression()
lm.fit(X_train1,y_train)
```

```python
lm.coef_ # coefficients of the features in the model
```

```python
lm = LinearRegression()
rfe = RFE(lm,n_features_to_select=10) # Recursive Feature Elimination (RFE) is a feature selection method that fits the model and removes the weakest features until the specified number of features is reached.
rfe = rfe.fit(X_train1, y_train)
```

```python
X_train1.columns[rfe.support_] # these are the features selected by RFE rfe.support means the features that are selected by RFE
```
