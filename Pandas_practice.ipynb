```python
import numpy as np
```

```python
temp = np.random.randint(low = 20, high = 100, size = [20,])
name = np.random.choice(['Abhay','Teclov','Geekshub','Aniket'],20)
random = np.random.choice([10,11,13,12,14],20)
```

```python
temp
```

```python
name
```

```python
random
```

```python
a = list(zip(temp, name, random)) # taking each value sequentially from the all arreys and making a list of arrays
a[:5]
```

```python
df = pd.DataFrame(data = a, columns=['temp','name','random']) #creating dataframe and naming columns
df
```

```python
df[:5] # this will give you a rows basis on index
```

```python
df["temp"] [:5] # how to extract single column data using indexing
```

```python
df["temp"].values # will give you values in array form
```

```python
df["temp"].max() # max value in temp column this type of functions can be only applied on series data
```

```python
df["temp"].min()
```

```python
df["temp"].sum()
```

```python
df["temp"].std()
```
