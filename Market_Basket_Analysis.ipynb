```python
# Grouping by 'Country' and summing 'Quantity'
data_plus[data_plus['Country'] =="United Kingdom"].groupby(['Invoice', 'Description'])['Quantity'].sum().unstack()
```

```python
# Creating a basket matrix for the United Kingdom
basket_plus = (data_plus[data_plus['Country'] =="United Kingdom"].groupby(['Invoice', 'Description'])['Quantity']
               .sum().unstack().reset_index().fillna(0).set_index('Invoice')) # unstack means conversion of intex to columns
basket_plus
```

```python
basket_plus.tail()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_encode_plus = basket_plus.applymap(encode_units)
basket_encode_plus
```

```python
basket_encode_plus.tail()
```

```python
basket_encode_plus > 0
```

```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(basket_filter_plus, min_support=0.03,
                                 use_colnames=True).sort_values
```

```python
apriori(basket_filter_plus, min_support=0.03,
                                 use_colnames=True).sort_values("support",ascending=False)
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.03,
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus.sort_values("length",ascending=False)
```

```python
frequent_itemsets_plus[ (frequent_itemsets_plus['length']  == 2) &
                        (frequent_itemsets_plus['support'] >= 0.03) ]
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets_plus, metric='lift',
                  min_threshold=1)
```

```python
association_rules(frequent_itemsets_plus, metric='lift',
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
